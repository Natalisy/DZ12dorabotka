Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

  если источник traffic_source равен yandex или google, то в source_type ставится organic
  для источников paid и email из России - ставим ad
  для источников paid и email не из России - ставим other
  все остальные варианты берем из traffic_source без изменений

```python
visit_log = pd.read_csv("visit_log.csv", sep = ';')
visit_log
```

```python
# функция для проверки источника траффика
def traffic_source_def(traffic):
    if traffic['traffic_source'] == 'yandex' or traffic['traffic_source'] == 'google':
        return 'organic'
    elif (traffic['traffic_source'] == 'paid' or traffic['traffic_source'] == 'email'):
        if traffic['region'] == 'Russia':
            return 'ad'
        else:
            return 'other'
    else:
        return traffic['traffic_source']
    
```

```python
# Применяем apply

visit_log['Source_type'] = visit_log.apply(traffic_source_def, axis = 1)
visit_log.head(30)
```

#### Задание 4
Дана статистика услуг перевозок клиентов компании по типам (см. файл с кодом занятия). 

rzd - железнодорожные перевозки
auto - автомобильные перевозки
air - воздушные перевозки
client_base - адреса клиентов

Необходимо сформировать две таблицы:
•	таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
•	аналогичную таблицу по типам выручки с указанием адреса клиента
Обратите внимание, что в процессе объединения таблиц данные не должны теряться.


```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
# Сливаем 3 таблицы одну по client_id

summer_table_id = rzd.merge(auto, how='outer')
summer_table_id = summer_table_id.merge(air, how='outer')
summer_table_id
```

```python
# Добавляем столбик с итоговой выручкой для каждого client_id

summer_table_id['total_revenues'] = summer_table_id.iloc[:, 1:].sum(axis=1)
summer_table_id
```

```python
# Сливаем в итоговую таблицу с адресами для каждого клиента

summer_table_address = summer_table_id.merge(client_base, how='outer')
summer_table_address
```
